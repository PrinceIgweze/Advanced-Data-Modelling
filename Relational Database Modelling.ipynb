## Database Normalization

The main activity of database normalization is to organize database into tables and columns
- Tables should be about a specific topic
- Only columns supporting that topic are included

Three reasons for normalization
- Minimize duplicate data
- Minimize or avoid data modification issues
- Simplify queries

The three normal forms
- 1st normal form
- 2nd normal form
- 3rd normal form
- Additional: Boyce-Codd

First Normal Form: The information is stored in a relational table and each column contains atomic values, and there are not repeating group of columns .<br>
Second Normal Form: The table is in the first normal form and all the columns depend on the tables's primary key.<br>
Third Normal Form: The table in the second normal form and all of its columns are not transitively dependent on the primary key.<br>

Type of Anomalies
- Update analomaly: In the context of database normalization, an update anomaly refers to problems that arise when updating data in a database that is not properly normalized. These anomalies can lead to inconsistencies, redundancy, and errors in the database. Heres a breakdown of what an update anomaly is and how it relates to normalization<br>

- Insertion Anomaly: This occurs when certain attributes cannot be inserted into the database without the presence of other attributes. For example, if you want to insert a new piece of information but cannot do so because it requires existing data, this is an insertion anomaly <br>

- Deletion Anomaly: This happens when deleting a piece of data inadvertently causes loss of other valuable information. For instance, if removing a record about a customer also removes all associated data about their purchases, even if the purchases are still relevant, this is a deletion anomaly <br>

- Modification Anomaly: This is the most directly related to the concept of update anomalies. It occurs when changing data in one place leads to inconsistencies because the same data is stored in multiple places. For example, if an address for a customer is updated in one record but not in others where the same customers information is stored, this can lead to inconsistent data.
